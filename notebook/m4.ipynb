```python
from value import Value
from nn import MLP
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("sweep.csv")
df_clean = df.dropna(subset=df.columns[6:13])
xs = df_clean.iloc[:, :6].values
ys = df_clean.iloc[:, 6:].values
df_clean
```

```python
print(np.isnan(xs).any(), np.isinf(xs).any())
print(np.isnan(ys).any(), np.isinf(ys).any())
```

```python
def balanced_sum(vals: list[Value]) -> Value:
    """
    Given a nonempty list of Value objects, return their sum
    in a balanced (â‰ˆlog-depth) tree.
    """
    if len(vals) == 1:
        return vals[0]
    mid = len(vals) // 2
    left = balanced_sum(vals[:mid])
    right = balanced_sum(vals[mid:])
    return left + right
  
n = MLP(6, [7, 7, 7])
lr = 1e-2

for k in range(1000):
    # 1) forward pass & collect all squared-errors in a list
    sq_errors = []  # type: list[Value]
    for x_row, y_row in zip(xs, ys):
        yp_vec = n(x_row.tolist())   # a list of 7 Values
        for yp_i, y_i in zip(yp_vec, y_row):
            sq_errors.append((yp_i - Value(float(y_i)))**2)

    # 2) build one loss Value via a balancedâ€tree sum
    loss = balanced_sum(sq_errors)

    # 3) backprop and update
    for p in n.parameters():
        p.grad = 0.0
    loss.backward()
    for p in n.parameters():
        p.data -= lr * p.grad

    print(f"step {k:2d}, loss = {loss.data:.6f}")
```
